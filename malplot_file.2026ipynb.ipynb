
## Matplotlib & Programming Definitions (Corrected & Improved)

**Data Point:** An individual value in a dataset, usually written as a coordinate (x, y).

**X-axis:** The horizontal axis that represents independent variables.

**Y-axis:** The vertical axis that represents dependent variables.

**Figure:** The complete drawing area or window that contains one or more plots.

**Axes:** The region inside the figure where data is plotted, including x and y axes.

**Plot:** A visual representation of data using lines, points, or other graphical elements.

**Line:** A graphical element that connects data points to show trends or relationships.

**Line Style:** The pattern used to draw a line (e.g., solid, dashed, dotted).

**Color:** The visual appearance assigned to lines, markers, or text in a plot.

**Legend:** A box that identifies and explains different elements in the plot.

**Labels:** Text used to describe axes (e.g., X-axis label, Y-axis label).

**Title:** A heading that describes the purpose or content of the graph.

**Grid:** Background reference lines that improve readability of the plot.

**Function:** A reusable block of code that performs a specific task.

**Method:** A function that is associated with an object and operates on it.

**Parameter:** A variable used in a function definition to accept input values.

**Keyword Argument:** A parameter passed to a function using its name (e.g., color='blue').

**Object-Oriented API:** A way of plotting using objects like Figure and Axes instead of simple functions.

**DPI (Dots Per Inch):** A measure of the resolution or clarity of a figure.

**Backend:** The system responsible for displaying or saving the graphical output.
